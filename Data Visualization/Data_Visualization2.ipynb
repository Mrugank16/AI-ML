```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
age_list = list(np.random.randint(20,40,12))
sal_list=list(np.random.randint(15,50,12))
emp={'Age':age_list,
     'Salary':sal_list}
emp_df=pd.DataFrame(emp)
emp_df
```

# Boxplot

```python
plt.figure(figsize=(5,6))
ax = sns.boxplot(data=emp_df)
plt.yticks(range(15,55,5))
plt.xlabel('Age & Salary')
plt.show()
```

```python
emp_df.plot(kind='box', figsize=(5,6), color='g', yticks=range(15,55,5))
```

# Outliers

```python
ser = pd.Series(sal_list)
ser
```

```python
q1 = ser.quantile(0.25)
q3 = ser.quantile(0.75)
iqr = q3-q1
print(q1)
print(q3)
print(iqr)
```

# Outliers Detection Formula

```python
upside = q3 + (1.5*iqr)
upside
```

```python
lowerside = q1-(1.5*iqr)
lowerside
```

```python
ser.plot(kind='box',figsize=(5,6), color='g', yticks=range(15,55,5))
```

```python
tips['time'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips, hue='time',style='smoker')
plt.show()
```
