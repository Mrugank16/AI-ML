# Types of Statistics

1. Descriptive --> applied on whole population
2. inferential --> applied on sample of whole population and assumed that the results produced by sample will also replicate the whole population

# Parameter and Statistics

```python
import statistics as st
import pandas as pd
import numpy as np
```

```python
x=[15,62,95,85,62,10,15,62,21,54]
print(st.mean(x))
print(st.median(x))
print(st.mode(x))
```

```python
x=pd.DataFrame(x)
print(x.mean())
print(x.median())
print(x.mode())
```

# Inferential Stats

1. Normal Distribution
  - probability bell curve
  - mean=0 and std=1

  1. Skewed Right
    - Mode>Median>mean
  2. Skewed Left
    - Mean>Median>Mode
2. Symmetric Distribution
  - perfect graph, well bell curve

3. Uniform Distribution
  - refers to probability distribution where all outcomes are equally likely

  

# Central Limit Theorem

- sample std = (population std/sqrt of sample size)
