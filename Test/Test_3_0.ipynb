Adding a Column (2 marks)

Matrix Operations (2 marks)
Create a 3x3 identity matrix using numpy.

```python
def create_identity(mat):
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            if i==j:
                mat[i][j]=1
            else:
                mat[i][j]=0
    return

mat=np.zeros((3,3))
create_identity(mat)
mat
```
